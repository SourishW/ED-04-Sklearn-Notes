```python
# importing necessary libraries
import matplotlib.pyplot as plt
```

```python
#first create x and y
x = [i for i in range(10)]
print(x)
```

```python
y = [2*i for i in range(10)]
print(y)
```

```python
#creating labels on x and y axis
plt.xlabel('x-axis')
plt.ylabel('y-axis')
#plotting the graph
plt.plot(x, y)
```

```python
plt.xlabel('x-axis')
plt.ylabel('y-axis')
# you can also put a scatter plot
plt.scatter(x, y)
```

```python
# you can even add colors
plt.scatter(x, y, color = 'r')
```
