```python
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt

# load the dataset
boston = datasets.load_boston()

# features and labels
X = boston.data
y = boston.target
```

```python
# take a look at things
print("x", X.shape, X)
print("y", y.shape, y)
```

```python
# now, to create the model/algorithm that is going to be used
l_reg = linear_model.LinearRegression()
```

```python
# now, let's visualize the data

# we need to do a matrix transpose to select which feature you want

plt.scatter(X.T[12],y, color='r')
# you can use this one with a polynomial regressor
```

```python
plt.scatter(X.T[2],y, color='g')
# maybe you can use a sigmoid graph with this?
```

```python
plt.scatter(X.T[10],y, color='b')
# maybe you can use a sigmoid graph with this?
```

```python
plt.scatter(X.T[5],y)
```

```python
# this one is the most linear one
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# training the model
```

```python
model = l_reg.fit(X_train, y_train)

predictions = model.predict(X_test)
```

```python
# testing the predictions
```

```python
print("Predictions: ", predictions)
```

```python
# R^2 value
print("R^2 value: ", l_reg.score(X, y))
```

```python
plt.scatter(predictions, y_test)
```

```python
# now, let's look at the coefficients, and the y intercept
print("coefficients: ", l_reg.coef_)
```

```python
print("intercept: ", l_reg.intercept_)
```
